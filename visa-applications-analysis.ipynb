```python
import plotly.graph_objects as go
import plotly.io as pio
```

# Preprocessing

```python
df = pd.read_csv(f'../input/visa-issuance-by-nationality-and-region-in-japan/{filename}')
df.head(20)
```

```python
total = df[df['Country']=='total']
total.head(15)
```

```python
df = df[df['Country']!='total'].reset_index(drop=True)
df.head(20)
```

# Visualize

## Purpose of visiting

```python
for purpose in df.columns[5:]:
    print(purpose)
```

## Total Number of Issued 

```python
x = total['Year']
s = 'Number of issued'
y = total[s]
plot = [
    go.Bar(
        x=x, y=y, name=s,
        marker_color='rgb(55, 83, 109)'
    )
]

layout = go.Layout(
    title='Total Number of Issued in Japan',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Issued Numbers',
        tick0=1000000,
        dtick=500000,
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```

## Top 10 Counties of Visa Issurance in Japan

```python
for year in df['Year'].unique():
    tmp = df[df['Year']==year].sort_values(by='Number of issued', ascending=False).head(20)
    x = tmp['Country']
    s = 'Number of issued'
    y = total[s]
    plot = [
        go.Bar(
            x=x, y=y, name=s,
            marker_color='rgb(55, 83, 109)'
        )
    ]

    layout = go.Layout(
        title=f'Top 10 Countries in {year}',
        xaxis_tickfont_size=14,
        yaxis=dict(
            title='Issued Numbers',
            tick0=1000000,
            dtick=500000,
            titlefont_size=16,
            tickfont_size=14,
        ),
        legend=dict(
            x=1.0, y=1.0,
            bgcolor='rgba(255, 255, 255, 0)',
            bordercolor='rgba(255, 255, 255, 0)'
        ),
        bargap=0.15,
        bargroupgap=0.1
    )

    fig = go.Figure(data=plot, layout=layout)
    fig.show()
```
