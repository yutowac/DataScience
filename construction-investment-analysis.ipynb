# Preprocessing

```python
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df = pd.read_csv(f'../input/construction-investment-amount-in-japan/{filename}')
df.head(20)
```

```python
df["Construction_Investment"].unique()
```

```python
df["Value"].unique()
```

```python
amounts = df.columns[4:]
amounts
```

```python
col_large = []
col_middle = []
col_small = []
for amount in amounts: 
    if "Total" in amount:
        pass
    else:
        n = amount.count('_')
        if n==1:
            col_large.append(amount)
        elif n==2:
            col_middle.append(amount)
        elif n==3:
            col_small.append(amount)
        else:
            print(amount)
```

```python
col_large
```

```python
col_middle
```

```python
col_small
```

```python
tmp = df[df["Construction_Investment"]=="amount"]
cols = col_large+["Year"]
tmp = tmp[cols]
tmp.head(15)
```

# Visualize

```python
x = tmp['Year']
plot = []
for col in col_large:
    plot.append(go.Bar(
        x=x, y=tmp[col], name=col
    ))

layout = go.Layout(
    title='Amount of Architecture Investment in Japan',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Amounts(100M/JPY)',
        tick0=100000,
        dtick=100000,
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1,
    barmode="relative" 
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```

* Proportion of amounts doesn't change by years until Building Repair added on the items.
