## 1. Number of Cases

```python
tmp = df[df['crime category']=='total number of crimes']
tmp.head(15)
```

```python
x = tmp['year']
s1 = 'reported cases'
y1 = tmp[s1]
s2 = 'cleared cases'
y2 = tmp[s2]
plot = [
    go.Bar(
        x=x, y=y1, name=s1,
        marker_color='rgb(55, 83, 109)'
    ),
    go.Bar(
        x=x, y=y2, name=s2,
        marker_color='rgb(26, 118, 255)'
    )
]

layout = go.Layout(
    title='Amount of Cases in Japan',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Cases',
        tick0=100000,
        dtick=100000,
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```

## 2. Cases by Crime Category

```python
tmp = df[df['crime category']!='total number of crimes']
tmp.head(15)
```

### 2-1. Sum of Cases by Each Crime Category

```python
y1_data = []
y2_data = []
for c_c in tmp['crime category'].unique():
    tmp2 = tmp[tmp['crime category']==c_c]
    y1_data.append(tmp2['reported cases'].sum())
    y2_data.append(tmp2['cleared cases'].sum())
```

```python

x = tmp['crime category'].unique()
s1 = 'reported cases'
s2 = 'cleared cases'
plot = [
    go.Bar(
        x=x, y=y1_data, name=s1,
        marker_color='rgb(55, 83, 109)'
    ),
    go.Bar(
        x=x, y=y2_data, name=s2,
        marker_color='rgb(26, 118, 255)'
    )
]

layout = go.Layout(
    title='Sum of Cases for 10 Years from 2006 to 2016',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Cases',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```

### 2-2. Each Crime Category with Error Bars

```python
y1_mean = []
y1_std = []
y2_mean = []
y2_std = []
for c_c in tmp['crime category'].unique():
    tmp2 = tmp[tmp['crime category']==c_c]
    y1_mean.append(tmp2['reported cases'].mean())
    y1_std.append(tmp2['reported cases'].std())
    y2_mean.append(tmp2['cleared cases'].mean())
    y2_std.append(tmp2['cleared cases'].std())
```

```python
fig = go.Figure()
x = tmp['crime category'].unique()
s1 = 'reported cases'
s2 = 'cleared cases'
plot = [
    go.Bar(
        x=x, y=y1_mean, name=s1,
        marker_color='rgb(55, 83, 109)',
        error_y=dict(type='data', array=y1_std)
    ),
    go.Bar(
        x=x, y=y2_mean, name=s2,
        marker_color='rgb(26, 118, 255)',
        error_y=dict(type='data', array=y2_std)
    )
]
layout = go.Layout(
    title='Average Cases for 10 Years from 2006 to 2016',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Average Cases',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1
)
fig = go.Figure(data=plot, layout=layout)
fig.show()
```

### 2-3. Cases Plot by Years

```python
years = tmp['year'].unique()
y1 = {}
y2 = {}
for year in years:
    y1_data = []
    y2_data = []
    for c_c in tmp['crime category'].unique():
        tmp2 = tmp[(tmp['crime category']==c_c)|(tmp['year']==year)|(tmp['crime name'].isnull())]
        y1_data.append(tmp2['reported cases'].sum())
        y2_data.append(tmp2['cleared cases'].sum())
    y1[year] = y1_data
    y2[year] = y2_data
```

```python
for year in years:
    x = tmp['crime category'].unique()
    s1 = 'reported cases'
    s2 = 'cleared cases'
    plot = [
        go.Bar(
            x=x, y=y1[year], name=s1,
            marker_color='rgb(55, 83, 109)'
        ),
        go.Bar(
            x=x, y=y2[year], name=s2,
            marker_color='rgb(26, 118, 255)'
        )
    ]

    layout = go.Layout(
        title=f'Cases in {year}',
        xaxis_tickfont_size=14,
        yaxis=dict(
            title='Cases',
            titlefont_size=16,
            tickfont_size=14,
        ),
        legend=dict(
            x=1.0, y=1.0,
            bgcolor='rgba(255, 255, 255, 0)',
            bordercolor='rgba(255, 255, 255, 0)'
        ),
        bargap=0.15,
        bargroupgap=0.1
    )

    fig = go.Figure(data=plot, layout=layout)
    fig.show()
```
