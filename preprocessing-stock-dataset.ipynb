# <div style="box-shadow: rgba(0, 0, 0, 0.16) 0px 1px 4px inset, rgb(51, 51, 51) 0px 0px 0px 3px inset; padding:20px; font-size:32px; font-family: consolas; text-align:center; display:fill; border-radius:15px;  color:rgb(34, 34, 34); background-color:rgb(255,255,255); "> <b> 2. Make Dataset🗃️ </b></div>

```python
# Tickers
GAFAM = ['GOOG', 'AAPL', 'META', 'AMZN', 'MSFT']
FAANG = ['META', 'AAPL', 'AMZN', 'NFLX', 'GOOG']
MATANA = ['MSFT', 'AAPL', 'TSLA', 'GOOG', 'NVDA', 'AMZN']
```

```python
d_GAFAM = {}
for ticker in GAFAM:
    stock = yf.Ticker(ticker) # access ticker data
    hist = stock.history(period="max")
    d_GAFAM[ticker] = hist
```

```python
d_FAANG = {}
for ticker in FAANG:
    stock = yf.Ticker(ticker) # access ticker data
    hist = stock.history(period="max")
    d_FAANG[ticker] = hist
```

```python
d_MATANA = {}
for ticker in MATANA:
    stock = yf.Ticker(ticker) # access ticker data
    hist = stock.history(period="max")
    d_MATANA[ticker] = hist
```

<a id="3"></a>
# <div style="box-shadow: rgba(0, 0, 0, 0.16) 0px 1px 4px inset, rgb(51, 51, 51) 0px 0px 0px 3px inset; padding:20px; font-size:32px; font-family: consolas; text-align:center; display:fill; border-radius:15px;  color:rgb(34, 34, 34); background-color:rgb(255,255,255); "> <b> 3. Save🛒 </b></div>

```python
import pickle
```

```python
with open("gafam_stock.pkl", "wb") as tf:
    pickle.dump(d_GAFAM, tf)
with open("faang_stock.pkl", "wb") as tf:
    pickle.dump(d_FAANG, tf)
with open("matana_stock.pkl", "wb") as tf:
    pickle.dump(d_MATANA, tf)
```
