```python
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as pc
```

```python
df1 = pd.read_csv(f'../input/consumer-price-index-of-japan-by-2022/{filenames[0]}')
print(df1.head(5))
df2 = pd.read_csv(f'../input/consumer-price-index-of-japan-by-2022/{filenames[1]}')
print(df2.head(5))
df3 = pd.read_csv(f'../input/consumer-price-index-of-japan-by-2022/{filenames[2]}')
print(df3.head(5))
df4 = pd.read_csv(f'../input/consumer-price-index-of-japan-by-2022/{filenames[3]}')
print(df4.head(5))
```

# General CPI trends

```python
df1.columns[1:]
```

```python
item1 = df1.columns[2:15]
item2 = df1.columns[15:33]
item3 = df1.columns[33:]
colors = pc.qualitative.Alphabet
```

```python
x = df1['Year']
plot = []
for i,col in enumerate(item1):
    color = colors[i]
    s = col
    y = df1[col]
    plot.append(
        go.Scatter(
            x=x, y=y, name=s,
            marker_color=color
        )
    )

layout = go.Layout(
    title='Consumer Price Index(CPI) in Japan',
    xaxis=dict(
        tickfont_size=14
    ),
    yaxis=dict(
        title='CPI',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1,
    colorscale=dict(sequential='spectral')
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```
