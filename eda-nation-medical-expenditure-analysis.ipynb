# üìèPreprocessing

```python
df = pd.read_csv(f'../input/nation-medical-expenditure-in-japan/{filename}')
df.head(20)
```

# üìàVisualize

```python
x = df['Year']
s1 = 'NationalMedicalExpenditure'
y1 = df[s1]
s2 = 'GrossDomesticProduct'
y2 = df[s2]
plot = [
    go.Bar(
        x=x, y=y1, name=s1,
        marker_color='rgb(55, 83, 109)'
    ),
    go.Bar(
        x=x, y=y2, name=s2,
        marker_color='rgb(26, 118, 255)'
    )
]

layout = go.Layout(
    title='Relation of National medical expenditure and Gross domestic products',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='100 million yen',
#         tick0=100000,
#         dtick=100000,
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```

```python
plot = go.Scatter(x=df[s1], y=df[s2],mode='markers')
layout = go.Layout(
    title='Relation of National medical expenditure and Gross domestic products',
    xaxis=dict(
        title=s1,
        titlefont_size=16,
        tickfont_size=14,
    ),
    yaxis=dict(
        title=s2,
        titlefont_size=16,
        tickfont_size=14,
    )
)
fig = go.Figure(data=plot, layout=layout)
fig.show()
```
