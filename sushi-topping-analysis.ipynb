```python
df = pd.read_csv(f'../input/seasonal-sushi/sushi.csv', encoding='shift-jis')
df.head(20)
```

```python
months = df.columns[5:17]
l_seasonal = []
toppings = {}
for month in months:
    l_seasonal.append(len(df[df[month]==1]))
    toppings[month] = list(df.loc[(df[month]==1),'Name_en'])
```

<a id="4"></a>
# <div style="box-shadow: rgba(0, 0, 0, 0.16) 0px 1px 4px inset, rgb(51, 51, 51) 0px 0px 0px 3px inset; padding:20px; font-size:32px; font-family: consolas; text-align:center; display:fill; border-radius:15px;  color:rgb(34, 34, 34);"> <b> 4. Visualize üìà</b></div>


## Seasonal Topping

```python
x = months
s = 'Number of Seasonals'
y = l_seasonal
plot = [
    go.Bar(
        x=x, y=y, name=s,
        marker_color='rgb(55, 83, 109)'
    )
]

layout = go.Layout(
    title='Number of Seasonal Sushi Toppings',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title=s,
        tick0=1,
        dtick=10,
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```
