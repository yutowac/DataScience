```python
import plotly.graph_objects as go
import plotly.io as pio
```

# Preprocessing

```python
df = pd.read_csv(f'../input/seasonal-sushi/{filename}')
df.head(20)
```

# Visualize

```python
months = df.columns[5:17]
l_seasonal = []
for month in months:
    l_seasonal.append(len(df[df[month]==1]))
```

## Seasonal Topping

```python
x = months
s = 'Number of Seasonals'
y = l_seasonal
plot = [
    go.Bar(
        x=x, y=y, name=s,
        marker_color='rgb(55, 83, 109)'
    )
]

layout = go.Layout(
    title='Number of Seasonal Sushi Toppings',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title=s,
        tick0=1,
        dtick=10,
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```
