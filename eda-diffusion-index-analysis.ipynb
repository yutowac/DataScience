# Preprocessing

```python
import plotly.graph_objects as go
import plotly.io as pio
from datetime import datetime
```

```python
df = pd.read_csv(f'../input/diffusion-index-in-japan/{filename}')
df.head(20)
```

```python
df["date"] = pd.to_datetime(df["year"].astype("str") + "/" + df["month"].astype("str") + "/" + df["day"].astype("str"))
df["date"]
```

# Visualize

## Total DI trend

```python
x = df['date']
col = "Total"
s1 = 'Current Conditions'
y1 = df[df["DI"]==s1][col]
s2 = 'Outlook'
y2 = df[df["DI"]==s2][col]
plot = [
    go.Scatter(
        x=x, y=y1, name=s1,
        marker_color='rgb(55, 83, 109)'
    ),
    go.Scatter(
        x=x, y=y2, name=s2,
        marker_color='rgb(26, 118, 255)'
    )
]

layout = go.Layout(
    title='Diffusion Index(DI) in Japan',
    xaxis=dict(
        tickfont_size=14
    ),
    yaxis=dict(
        title='DI',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```

## Each Component of DI 

```python
tmp = df[df["DI"]=="Current Conditions"]
x = tmp['date']
s1 = 'HouseholdTrend'
y1 = tmp[s1]
s2 = 'CorporateTrend'
y2 = tmp[s2]
s3 = 'Employment'
y3 = tmp[s3]

plot = [
    go.Scatter(
        x=x, y=y1, name=s1,
        marker_color='rgb(55, 83, 109)'
    ),
    go.Scatter(
        x=x, y=y2, name=s2,
        marker_color='rgb(26, 118, 255)'
    ),
    go.Scatter(
        x=x, y=y3, name=s3,
        marker_color='rgb(34, 140, 111)'
    )
]

layout = go.Layout(
    title='Diffusion Index(DI) in Japan',
    xaxis=dict(
        tickfont_size=14
    ),
    yaxis=dict(
        title='DI',
        titlefont_size=16,
        tickfont_size=14,
    ),
    legend=dict(
        x=1.0, y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),
    bargap=0.15,
    bargroupgap=0.1
)

fig = go.Figure(data=plot, layout=layout)
fig.show()
```

# Conclusion

- Current Conditions and Outlook are the almost same trend by years.
